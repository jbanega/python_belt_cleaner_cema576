**CEMA 576. Classification of Applications for Bulk Materials Conveyor Belt Cleaning**

```python
import ipywidgets as widgets
from ipywidgets import Layout
style = {'description_width': 'initial'}
from IPython.display import Image
Image("img/Belt-Cleaner-Models.png")
```

**INTRODUCTION**

*The proper selecion of a belt cleaner must consider the environmentin wich the belt cleaner must operate. Several factors will play significant roles in deciding the appropiate selection. This standart provides a method to condense a complex operating environment into a single classification number to be used when selecting belt cleaners.*

**RANKINGS AND CLASS SYSTEM**

*The application class is determined using the following factors:*
1. Belt width,
2. Belt speed.
3. Type of belt splices.
4. Abrasiveness of the material.
5. Stickness/ moisture content of the material.

*For the purposes of this classification conveyor belts must be new or in "as new" conditions*

**BELT CHARACTERISTICS**

**Belt width**

```python
Belt_width = widgets.RadioButtons(
    options=['<24 in (<610 mm)',
             '24-42 in (610-1067 mm)',
             '42-60 in (1068-524 mm)',
             '60-96 in (1525-2438 mm)',
             '>96 in (>2438 mm)'],
    description='Belt width:',
    disabled=False
)    
Belt_width
```

**Belt speed**

```python
Belt_speed = widgets.RadioButtons(
    options=['<300 fpm (<1.5 m/s)',
             '300-600 fpm (1.5-3.0 m/s)',
             '601-1000 fpm (3.1-5.0 m/s)',
             '>1000 fpm (> 5.0 m/s)'],
    description='Belt speed:',
    disabled=False,
    layout=Layout(width='50%', height='80px'),
    style=style
)
Belt_speed
```

**Splice Type**

```python
Splice_type = widgets.RadioButtons(
    options=['vulcanized (for the entire life of the belt)',
             'Mechanical splices with belt speed below 500 fpm (<2.5 m/s)',
             'Mechanical splices with belt speed 500 fpm or greater (= or >2.5 m/s)'],
    description='Splice_type:',
    disabled=False,
    layout=Layout(width='70%', height='80px'),
    style=style
)
Splice_type
```

**MATERIAL CHARACTERISTICS**

**Abrasiveness**

```python
Abrasiveness = widgets.RadioButtons(
    options=['Mild   (CEMA Standard 550 code 5, abrasive index 1-17)',
             'Medium (CEMA Standard 550 code 6, abrasive index 18-67)',
             'Severe (CEMA Standard 550 code 7, abrasive index 68-416)'],
    description='Abrasiveness:',
    text="""NOTE: Choose the worst case expected conditions in situations where the conditions will vary""",
    disabled=False,
    layout=Layout(width='50%', height='80px'),
    style=style
)
Abrasiveness
```

**Stickness/ Moisture content**

```python
Stickness = widgets.RadioButtons(
    options=['Mild/ Dry     (<2% moisture by weight)',
             'Medium/ Moist (2-8% moisture by weight)',
             'Heavy/ Wet    (>8% moisture by weight)',
             'Severe/ Wet slurry with fines (CEMA 550 code F, O, V)'],
    description='Stickness:',
    disabled=False,
    layout=Layout(width='50%', height='80px'),
    style=style
)
Stickness
```

**APPLICATION SEVERITY RANKING**

```python
# Belt width
if Belt_width.value == '<24 in (<610 mm)':
    var1 = 0
elif Belt_width.value == '24-42 in (610-1067 mm)':
    var1 = 1
elif Belt_width.value == '42-60 in (1068-524 mm)':
    var1 = 2
elif Belt_width.value == '60-96 in (1525-2438 mm)':
    var1 = 4
elif Belt_width.value == '>96 in (>2438 mm)':
    var1 = 8

# Belt speed 
if Belt_speed.value == '<300 fpm (<1.5 m/s)':
    var2 = 1
elif Belt_speed.value == '300-600 fpm (1.5-3.0 m/s)':
    var2 = 2
elif Belt_speed.value == '601-1000 fpm (3.1-5.0 m/s)':
    var2 = 4
elif Belt_speed.value == '>1000 fpm (> 5.0 m/s)':
    var2 = 8

# Splice types
if Splice_type.value == 'vulcanized (for the entire life of the belt)':
    var3 = 0
elif Splice_type.value == 'Mechanical splices with belt speed below 500 fpm (<2.5 m/s)':
    var3 = 2
elif Splice_type.value == 'Mechanical splices with belt speed 500 fpm or greater (= or >2.5 m/s)':
    var3 = 4

# Abrasiveness
if Abrasiveness.value == 'Mild   (CEMA Standard 550 code 5, abrasive index 1-17)':
    var4 = 1
elif Abrasiveness.value == 'Medium (CEMA Standard 550 code 6, abrasive index 18-67)':
    var4 = 2
elif Abrasiveness.value == 'Severe (CEMA Standard 550 code 7, abrasive index 68-416)':
    var4 = 3

# Splice types
if Stickness.value == 'Mild/ Dry     (<2% moisture by weight)':
    var5 = 1
elif Stickness.value == 'Medium/ Moist (2-8% moisture by weight)':
    var5 = 2
elif Stickness.value == 'Heavy/ Wet    (>8% moisture by weight)':
    var5 = 4
elif Stickness.value == 'Severe/ Wet slurry with fines (CEMA 550 code F, O, V)':
    var5 = 8

TOTAL = var1 + var2 + var3 + var4 + var5
print('Aplication severity ranking: ' + str(TOTAL))
```

**BELT CLEANER CLASS**

```python
if TOTAL <= 6:
    print('You need a belt cleaner CEMA CLASS 1.')
elif TOTAL > 7 & TOTAL <10:
    print('You need a belt cleaner CEMA CLASS 2.')      
elif TOTAL > 11 & TOTAL <15:
    print('You need a belt cleaner CEMA CLASS 3.')
elif TOTAL > 16 & TOTAL <23:
    print('You need a belt cleaner CEMA CLASS 4.')
elif TOTAL > 24:
    print('You need a belt cleaner CEMA CLASS 5.')
```
